# minst hand written digits recognition 

```python
from __future__ import print_function
import numpy as np
from keras.datasets import mnist 
from keras.models import Sequential
from keras.layers.core import Dense , Activation,Dropout
from keras.optimizers import SGD,RMSprop , Adam
from keras.utils import np_utils
np.random.seed(1671) # for reproducibility
```

#### load dataset

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

#### reshape the dataset to make images flat


```python
RESHAPED = 784
#
X_train = X_train.reshape(60000, RESHAPED)
X_test = X_test.reshape(10000, RESHAPED)
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
```

#### noramlize images

```python
X_train /= 255
X_test /= 255
print(X_train.shape[0], 'train samples')
print(X_test.shape[0], 'test samples')
```

#### let's define our model hyper paramters

```python
NB_EPOCH = 200
BATCH_SIZE = 128
VERBOSE = 1
NB_CLASSES = 10   # number of outputs = number of digits
OPTIMIZER = SGD() # SGD optimizer, explained later in this chapter
N_HIDDEN = 128
VALIDATION_SPLIT=0.2 # how much TRAIN is reserved for VALIDATION
```

#### convert target values to OHE 

```python
Y_train = np_utils.to_categorical(y_train, NB_CLASSES)
Y_test = np_utils.to_categorical(y_test, NB_CLASSES)
```

#### define our model architecture

```python
model = Sequential()
model.add(Dense(NB_CLASSES, input_shape=(RESHAPED,)))
model.add(Activation('softmax'))
```

```python
model.summary()
```

#### compile our keras model 

```python
model.compile(loss='categorical_crossentropy',
              optimizer=OPTIMIZER,
              metrics=['accuracy'])
```

```python
history = model.fit(X_train, Y_train,
                    batch_size=BATCH_SIZE, epochs=NB_EPOCH,
                    verbose=VERBOSE, validation_split=VALIDATION_SPLIT)
```

#### evaluate our model

```python
score = model.evaluate(X_test, Y_test, verbose=VERBOSE)
print("\nTest score:", score[0])
print('Test accuracy:', score[1])
```

# let's improve this model by adding more hidden layers

```python
### i will modifiy model architecture only
model = Sequential()
model.add(Dense(128,input_shape=(784,)))
model.add(Activation('relu'))
model.add(Dense(128))
model.add(Activation('relu'))
model.add(Dense(10))
model.add(Activation('softmax'))

model.summary()
```

```python
model.compile(loss='categorical_crossentropy',optimizer=OPTIMIZER,metrics=['accuracy'])
model.fit(X_train,Y_train,epochs=NB_EPOCH,
          batch_size=BATCH_SIZE,verbose=True,
          validation_split=0.2)
```

##### evaluation after improvments

```python
score = model.evaluate(X_test,Y_test)
print("\nTest score:", score[0])
print('Test accuracy:', score[1]*100)
```

Now we reached 97.6 % accuracy on test set . but there are more improvements ? , let's see

let's add some dropout to prevent overfitteing 

```python

model = Sequential()
model.add(Dense(128,input_shape=(784,)))
model.add(Activation('relu'))
model.add(Dropout(.3)) # drop .3 of connections between layers

model.add(Dense(128))
model.add(Activation('relu'))
model.add(Dropout(.3))

model.add(Dense(10))
model.add(Activation('softmax'))

model.summary()
```

```python
model.compile( loss='categorical_crossentropy',
              optimizer=OPTIMIZER,metrics=['accuracy'])

model.fit(X_train,Y_train, epochs=100 , batch_size=BATCH_SIZE 
           , verbose=True , validation_split=.2)
```

```python
# evaluations 
score = model.evaluate(X_test,Y_test)
print ("test score",score[0])
print ("test accuracy",score[1]*100)
```

# trying different optimizers

use the previous model with differenet optimizers

```python
score = model.evaluate(X_test,Y_test)
print ("test score ",score[0])
print ("test accuracy",score[1]*100)
```
